```python
# import iris as iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn

sbn.set(style="white", color_codes=True)
% matplotlib inline
```

```python
df = pd.read_csv("./Iris.csv")
```

```python
df.head()
```

```python
df.plot(kind='scatter', x="SepalLengthCm", y='SepalWidthCm');
```

```python
sbn.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df);
```

```python
sbn.FacetGrid(df, hue='Species', size=5).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend();
```

```python
sbn.boxplot(x='Species', y='PetalLengthCm', data=df);
```
